## train_csv 데이터 받아오기

## x_train과 y_train 선언

### 선언 후, x_train과 y_train이 잘 나왔는지 확인

```python
x_train = data_train.drop(["Activity"], axis = 1)
y_train = data_train["Activity"]
x_train
```

```python
y_train
```

## Pipeline 및 StandardScaler사용

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

num_pipline = Pipeline([
    ('std_scaler',StandardScaler()),
    ('Imputer', SimpleImputer(strategy="median"))
])
x_train = num_pipline.fit_transform(x_train)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix
```

## SVC(C 값을 조기종료 방식으로 찾기)

### cv10이란 함수 선언(cv10을 이용해 비교하여 더 내려갈 때, C값 찾기)

```python
def cv10(model, x_train, y_train):
    cv10_val_score = cross_val_score(model,x_train,y_train, scoring="accuracy", cv=10)
    cv10_score = np.mean(cv10_val_score)
    return cv10_score
```

```python
svm_clf = SVC(kernel = 'rbf', random_state=42)

best_cv10_score = 0
error_going_up = 0
best_C = 0
for C in range(10,1000,1):
    C = C / 10;
    svm_clf.C = C
    svm_clf.fit(x_train, y_train)
    scores_checking = cv10(svm_clf,x_train, y_train)
    if scores_checking > best_cv10_score:
        best_cv10_score = scores_checking
        error_going_up = 0
        print(f"cv : {best_cv10_score}, C: {C}")
        best_C = C
    else:
        error_going_up += 1
        print(f"over the best_cv!! {scores_checking} C: {C}, best_C: {best_C} ,error_count: {error_going_up}")
        if error_going_up == 30:
            svm_clf.C = best_C
            break
```

```python
best_C
```

```python
svm_best_clf = SVC(kernel = 'rbf', C = best_C, random_state=42, probability = True)
svm_best_clf.fit(x_train, y_train)
```

## Adaboost (base_estimator = DecisionTreeClassifier)

###### Adaboosting 역시 조기종료방식(n_estimators을 찾는 방식)을 사용했으나, Desktop에서 24시간 넘게 돌려도 안끝나서 계산된 것 중 최상의 값을 넣었습니다.

```python
ada_best_clf = AdaBoostClassifier(base_estimator = DecisionTreeClassifier(max_depth=10, 
                                                                      min_samples_split=3, random_state=42),
                              random_state=42, n_estimators = 160)
```

## 각각의 모델들의 K-fold의 평균 값 출력

```python
model = (svm_best_clf,ada_best_clf)
for x in model:
    test = cv10(x,x_train,y_train)
    print(f"model : {x}, cv: {test}")
```

* SVC : 0.9892542886719905
* Ada : 0.990342539189589

## Voting(SVC, Ada)

#### Voting 방식 : soft

```python
vot_clf_best = VotingClassifier(
    estimators=[('svm', svm_best_clf), ('ada',ada_best_clf)], 
    voting = 'soft', n_jobs = -1)
vot_clf_best.fit(x_train, y_train)
```

## cv10 이후 값 출력

```python
score_best = cross_val_score(vot_clf_best, x_train, y_train, cv=10, scoring="accuracy")
score_best
```
